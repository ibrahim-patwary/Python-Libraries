# Create Dataframe

```python
import pandas as pd
import numpy as np
dates = pd.date_range("today",periods=7) #Define time sequence as index
num_ar= np.random.randn(7,4) #Import numpy random array
columns =['A','B','C','D'] #Use the table as the column name
df = pd.DataFrame(num_ar , index = dates , columns =columns)
print(df)
```

# Remove Repeated or duplicates data using pandas
