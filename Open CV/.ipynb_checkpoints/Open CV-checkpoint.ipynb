# Displaying Text

```python
image_path = r"Car_Picture.jpg"
image= cv.imread(image_path)
text ="Rolls Royce"
coordinates=(230,50)
font=cv.FONT_HERSHEY_COMPLEX_SMALL
fontScale=2
color =(250,25,20)
thickness =2
display = cv.putText(image,text,coordinates,font,fontScale,color,thickness)

plt.imshow(display)
plt.show()
```

# Reading and Displaying an Image

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img= cv.imread("cat_1.jpg")
cv.imshow("orginal picture",img)

cv.waitKey(0)
cv.destroyAllWindows()
```

# Reding Videos

```python
capture = cv.VideoCapture("short.mp4")

while True:
    isTrue, frame = capture.read()
    if not isTrue:
        break
        
    cv.imshow("Orginal Video", frame)
    
    if cv.waitKey(20) & 0xFF==ord("q"):
        break
        
capture.release()

cv.waitKey(0)
cv.destroyAllWindows()
```

# Changing Color-Space

```python
img= cv.imread("cat_1.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow("change picture",gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Edge Cascade

```python
img= cv.imread("Car_Picture.jpg")
canny = cv.Canny(img, 125,175)
cv.imshow("Canny Edges",canny)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Cropping the Image

```python
img= cv.imread("cat_1.jpg")
cv.imshow("orginal picture",img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Cropping
cropped = img[60:260,200:600]
cv.imshow("Cropped picture",cropped)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Flipping

```python
img= cv.imread("cat_1.jpg")
flip= cv.flip(img,-5)
cv.imshow("Flip",flip)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Thresh

```python
img= cv.imread("Car_Picture.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow("change picture",gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
ret , thresh = cv.threshold(gray,125,255,cv.THRESH_BINARY)
cv.imshow("Thresh picture",thresh)
cv.waitKey(0)
cv.destroyAllWindows()
```
